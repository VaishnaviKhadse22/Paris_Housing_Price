```python
x = df.drop(columns=["price"])
y = df['price']
```

### Splitting Train-Test

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x,train_y,test_y = train_test_split(x,y,random_state=62)
```

### Data Model

```python
from sklearn.linear_model import LinearRegression
```

```python
mod_lr = LinearRegression()
```

```python
mod_lr.fit(train_x,train_y)
```

```python
y_pred=mod_lr.predict(test_x)
```

### Importing r2_score to check the accuracy of the model

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,test_y)*100
```

### Predicting the prices by giving input features to the model

```python
def predict_house_price(input_features):
    input_df = pd.DataFrame([input_features])
    predicted_price = mod_lr.predict(input_df)
    return predicted_price[0] 
```

```python
input_features = {'squareMeters': 4000, 'numberOfRooms':3, "hasYard":1, "hasPool":0,
                  "floors":2, "cityCode":27939, "cityPartRange":3, "numPrevOwners":0,
                  "made":2006, "isNewBuilt":1, "hasStormProtector":0, "basement":0, 
                  "attic":0,"garage":0, "hasStorageRoom":1,"hasGuestRoom":0}

predicted_price = predict_house_price(input_features)
print(f"Predicted Price for the input features: {predicted_price}")
```
